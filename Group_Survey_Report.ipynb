# Spending time on each responsibility

The first section we are analyzing is our participants how much time they spend on each responsibility.
We use a radar graph to illustrate the distribution of participants' time spent on work across different academic years. The graph utilizes a radial chart to showcase the varying commitment levels for each specified time category, ranging from '0 hours' to '40+ hours.' The most significant observations are in Year 3 participants , indicating a substantial increase in participants spending time on work. Noteworthy concentrations are observed in the '8 - 16 hours' and '16 - 24 hours' categories. Year 4 participants present a pattern similar to Year 2, with a low overall involvement in work-related responsibilities.


```python
# Data for the radar chart
labels = ['0 hours', 
          'less than 8',  
          '8 - 16 hours', 
          '16 - 24 hours',
          '24 - 32 hours',
          '32 - 40 hours',
          '40+ hours']
year0 = [1, 0, 0, 0, 1, 0, 0]  #values for Year 0
year1 = [4, 2, 1, 3, 2, 1, 1]  #values for Year 1
year2 = [2, 2, 0, 1, 0, 0, 0]  #values for Year 2
year3 = [11, 1, 14, 7, 0, 1, 2]  #values for Year 3
year4 = [2, 1, 0, 1, 0, 0, 0]  #values for Year 4

# Number of variables
num_vars = len(labels)

# Compute angle for each axis
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# The plot is a circle, so we need to "complete the loop"
year0 += year0[:1]
angles += angles[:1]

# Plotting the data for each label
fig, ax = plt.subplots(figsize=(7, 7), subplot_kw=dict(polar=True))

ax.fill(angles, year0, color='skyblue', alpha=0.5)
ax.plot(angles, year0, color='blue', linewidth=2, label='Year 0')  # Outline

year1 += year1[:1]
ax.fill(angles, year1, color='salmon', alpha=0.5)
ax.plot(angles, year1, color='red', linewidth=2, label='Year 1')  # Outline

year2 += year2[:1]
ax.fill(angles, year2, color='lightgreen', alpha=0.5)
ax.plot(angles, year2, color='green', linewidth=2, label='Year 2')  # Outline

year3 += year3[:1]
ax.fill(angles, year3, color='orange', alpha=0.5)
ax.plot(angles, year3, color='darkorange', linewidth=2, label='Year 3')  # Outline

year4 += year4[:1]
ax.fill(angles, year4, color='plum', alpha=0.5)
ax.plot(angles, year4, color='purple', linewidth=2, label='Year 4')  # Outline

# Labels for each axis
ax.set_yticklabels([])
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Title
plt.title('In the past year, how much time did you spend on work?')

# Show the plot with legend
plt.legend(loc='center left', bbox_to_anchor=(1.5, 0.5))
plt.show()
```

We created the above radar chart to show how many hours each year group, spent on each area of responsibility (work, caring, volunteering, Extra-Curricular Activities, Societies and clubs, and other responsibilities). However, that because of the extreme values for students in year 3, this made the chart look very skewed and hard to read. This is was not a good way to visually communicate the data we wanted to show. We then chose to break the areas up and represent them as bar charts individually.

```python
hours = ['0 hours', 'less than 8', '8 - 16', '16 - 24', '24 - 32', '32 - 40', '40+']
year0 = [1, 0, 0, 0, 1, 0, 0]  #values for Year 0
year1 = [4, 2, 1, 3, 2, 1, 1]  #values for Year 1
year2 = [2, 2, 0, 1, 0, 0, 0]  #values for Year 2
year3 = [11, 1, 14, 7, 0, 1, 2]  #values for Year 3
year4 = [2, 1, 0, 1, 0, 0, 0]  #values for Year 4

# Number of responsibilities
num_hours = len(hours)

# Arrange the x-axis positions for bars
ind = np.arange(num_hours)

# Width of each bar
width = 0.15

# Plotting the bars for each year
plt.figure(figsize=(10, 6))

plt.bar(ind, year0, width, label='Year 0')
plt.bar(ind + width, year1, width, label='Year 1')
plt.bar(ind + 2 * width, year2, width, label='Year 2')
plt.bar(ind + 3 * width, year3, width, label='Year 3')
plt.bar(ind + 4 * width, year4, width, label='Year 4')

# X-axis labels and tick positions
plt.ylabel('Number of People')
plt.title('In the past year, how many hours per week have you spent on work?')
plt.xticks(ind + 2 * width, hours)
plt.legend()

# Show plot
plt.show()
```

This bar chart shows us that reveals significant variation in the time participants dedicate to caring responsibilities across academic years.Year 3 participants stand out with a higher level of no engagement in caring responsibilities, as evidenced by a substantial number reporting 0 hours per week.
Participants in Years 0, 2, and 4 consistently report spending fewer hours on caring responsibilities, with a concentration in the "less than 8" hours category. Also, most of our participants spent a lot of time on working  compare with caring.

## Word Count

The following code will count the number of words in Markdown cells. Code cells are not included.

- `Main word count` is the number of words in the main body of the text, *excluding* references or appendices.
- `References and appendices word count` is the number of words in any references or appendices.

Only `Main word count` relates to the assignment word limit. There is no limit to the number of words that can be included in references or appendices. Please note that appendices should only be used to provide context or supporting information. *No marks will be directly awarded for material submitted in appendices*.

Important:

- Please do not modify the word count code!
- To exclude references from your word count **you must** have a cell that starts with the text `## References`. Everything below this cell will not count towards the main word count.
- If you are submitting additional material as appendices **you must** have a cell that starts with the text `## Appendices`. Everything below this cell will not count towards the main word count. If you do not have any appendices you can delete the `## Appendices` cell.
- Code comments should only be used to explain details of the implementation, not for discussing your findings. All analysis commentary **must** be written in Markdown cells. *No marks will be awarded for analysis discussion submitted as comments in code cells*.

```python
%%js

// Run this cell to update your word count.

function wordcount() {
    let wordCount = 0
    let extraCount = 0
    let mainBody = true

    let cells = Jupyter.notebook.get_cells()
    cells.forEach((cell) => {
        if (cell.cell_type == 'markdown') {
            let text = cell.get_text()
            // Stop counting as main body when get to References or Appendices.
            if (text.startsWith('## References') ||
                text.startsWith('## Appendices')) {
                mainBody = false
            }
            if (text.startsWith('## Word Count')) {
                text = ''
            }
            if (text) {
                let words = text.toLowerCase().match(/\b[a-z\d]+\b/g)
                if (words) {
                    let cellCount = words.length
                    if (mainBody) {
                        wordCount += cellCount
                    } else {
                        extraCount += cellCount
                    }
                }
            }
        }
    })
    return [wordCount, extraCount]
}

let wc = wordcount()
element.append(`Main word count: ${wc[0]} (References and appendices word count: ${wc[1]})`)
```
